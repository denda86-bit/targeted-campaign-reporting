```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import datetime 
import numpy as np
```

```python
fig1, ax1 = plt.subplots()
labels= ["Direct", "Organic Search", "Referral", "Social"]
ax1.pie([34.6, 42, 15.2, 8.2], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
 # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.axis('equal')

plt.show()
plt.savefig('graph_brand1.png')
```

```python
todays_date = datetime.date(2019, 4, 15)
index = pd.date_range(todays_date, periods=20, freq='D')
index = [t.strftime('%d-%m-%Y') for t in index]

df2 = pd.DataFrame(index=index)
email = [random.randint(10, 1500) for x in range(len(df2.index))]
email.sort()
df2['Email'] = email

youtube = [random.randint(1, 50) for x in range(len(df2.index))]
youtube.sort()
df2['Youtube'] = youtube

facebook = [random.randint(10, 1000) for x in range(len(df2.index))]
facebook.sort()
df2['Facebook'] = facebook

twitter = [random.randint(10, 700) for x in range(len(df2.index))]
twitter.sort()
df2['Twitter'] = twitter

linkedin = [random.randint(10, 1000) for x in range(len(df2.index))]
linkedin.sort()
df2['LinkedIn'] = linkedin

df2 = df2[["Email", 'Facebook', 'LinkedIn', 'Twitter', 'Youtube']]
```

```python
ax = df2.plot(kind='bar', stacked=True, figsize=(12,8), title='Marketing Reach per Channel')
ax.set_ylabel("Number of subscribers/followers ")
plt.savefig('graph_brand2.png')
```

```python
df2['Impressions'] = [random.randint(100, 700) for x in range(len(df2.index))] 
df2['Clicks'] = df2['Impressions'].apply(lambda x: x*random.uniform(0.1, 0.4))
df2['CTR'] = df2['Clicks']/df2['Impressions']*100

# df1 = pd.DataFrame()
# r = np.linspace(1, 5, 10)
# r = np.sort(-r)
# linear = r**2
# df1['crt'] = linear
# df1['crt'].head()
# df1.plot(style='.-', figsize=(8,5), legend=False, color="green")
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))
ax2 = ax1.twinx()  # set up the 2nd axis
df2[['Impressions', 'Clicks']].plot.bar(ax=ax1)
ax1.grid(b=False)
ax2.plot(df2['CTR'].values, color='black')
ax1.set_title('Click-Through-Rate')
ax1.set_ylabel('Daily number of impressions (blue bars) and \n clicks (orange bars)')
ax2.set_ylabel('Daily Click-Through-Rate [%]')
plt.savefig('graph_brand3.pdf')
```
